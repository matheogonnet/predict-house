```python
# Importing necessary libraries for visualization and analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Display plots inline in the notebook
%matplotlib inline

# Set a consistent aesthetic style for plots
sns.set_theme(style="whitegrid")
```

```python
# Load saved predictions and metrics for regression and classification models
regression_results = pd.read_csv("../results/regression/regression_predictions.csv")  
classification_results = pd.read_csv("../results/classification/classification_predictions.csv") 

# Display the first few rows of each dataset to understand their structure
print("Regression Results Sample:")
display(regression_results.head())

print("\nClassification Results Sample:")
display(classification_results.head())
```

```python
# Visualizing how well the regression models performed
plt.figure(figsize=(14, 6))

# Linear Regression scatter plot
plt.subplot(1, 2, 1)
plt.scatter(regression_results['Actual'], regression_results['Linear_Predictions'], alpha=0.6, edgecolor='k', s=50, color="skyblue")
plt.plot([regression_results['Actual'].min(), regression_results['Actual'].max()],
         [regression_results['Actual'].min(), regression_results['Actual'].max()], color='red', linestyle='--', linewidth=2)
plt.title("Linear Regression: Predictions vs Actual", fontsize=14)
plt.xlabel("Actual Values", fontsize=12)
plt.ylabel("Predicted Values", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Random Forest scatter plot
plt.subplot(1, 2, 2)
plt.scatter(regression_results['Actual'], regression_results['RF_Predictions'], alpha=0.6, edgecolor='k', s=50, color="lightcoral")
plt.plot([regression_results['Actual'].min(), regression_results['Actual'].max()],
         [regression_results['Actual'].min(), regression_results['Actual'].max()], color='red', linestyle='--', linewidth=2)
plt.title("Random Forest: Predictions vs Actual", fontsize=14)
plt.xlabel("Actual Values", fontsize=12)
plt.ylabel("Predicted Values", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Load feature importances for Random Forest
rf_feature_importances = pd.read_csv("../results/feature_importances/rf_feature_importances.csv")

# Plotting feature importances for Random Forest Regression
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=rf_feature_importances, palette="viridis")
plt.title("Feature Importance: Random Forest Regressor", fontsize=14)
plt.xlabel("Importance Score", fontsize=12)
plt.ylabel("Feature", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Generate confusion matrices for classification models
knn_cm = confusion_matrix(classification_results['Actual'], classification_results['KNN_Predictions'])
dt_cm = confusion_matrix(classification_results['Actual'], classification_results['DT_Predictions'])

# Plotting the confusion matrices
plt.figure(figsize=(14, 6))

# KNN Confusion Matrix Heatmap
plt.subplot(1, 2, 1)
sns.heatmap(knn_cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Low', 'Medium', 'High', 'Very High'], 
            yticklabels=['Low', 'Medium', 'High', 'Very High'], cbar=False)
plt.title("KNN Confusion Matrix", fontsize=14)
plt.xlabel("Predicted Class", fontsize=12)
plt.ylabel("Actual Class", fontsize=12)

# Decision Tree Confusion Matrix Heatmap
plt.subplot(1, 2, 2)
sns.heatmap(dt_cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Low', 'Medium', 'High', 'Very High'], 
            yticklabels=['Low', 'Medium', 'High', 'Very High'], cbar=False)
plt.title("Decision Tree Confusion Matrix", fontsize=14)
plt.xlabel("Predicted Class", fontsize=12)
plt.ylabel("Actual Class", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Load regression metrics
regression_metrics = pd.read_csv("../results/regression/regression_metrics.csv")

# Plotting R scores for regression models
plt.figure(figsize=(8, 5))
sns.barplot(data=regression_metrics, x='Model', y='R', palette="viridis")
plt.title("R Scores for Regression Models", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("R Score", fontsize=12)
plt.ylim(0, 1)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Load classification metrics
classification_metrics = pd.read_csv("../results/classification/classification_metrics.csv")

# Plotting accuracy scores for classification models
plt.figure(figsize=(8, 5))
sns.barplot(data=classification_metrics, x='Model', y='Accuracy', palette="viridis")
plt.title("Accuracy Scores for Classification Models", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.ylim(0, 1)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
